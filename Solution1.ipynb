```python
path =  "challenge/"
filename1 = "data_1__.csv"
filename2 = "data_2__.csv"
filename3 = "data_3__.csv"
filename4 = "data_4__.csv"
file1 = path + str(filename1)
file2 = path + str(filename2)
file3 = path + str(filename3)
file4 = path + str(filename4)
df_file1 = pd.read_csv(file1, delimiter=",")
df_file2 = pd.read_csv(file2, delimiter=",")
df_file3 = pd.read_csv(file3, delimiter=",")
df_file4 = pd.read_csv(file4, delimiter=",")
plt.rcParams['figure.figsize'] = (20, 5)
```

```python
model1 = Pipeline([('poly', PolynomialFeatures(degree=3)),
                   ('linear', LinearRegression(fit_intercept=False))])
```

```python
fat_test = df_file1['valor'].values[-1]
df_mfile1 = df_file1.drop(df_file1.index[-1])
fat_test = [[fat_test]]
```

```python
x = pd.DataFrame(df_file1['index'])
y = pd.DataFrame(df_file1['valor'])
model1.fit(x,y)
predf1 = (model1.predict([[588]]))
predf1
```

```python
model2 = LinearRegression()
model2.fit(x,y)
pred2f1 = (model2.predict([[588]]))
pred2f1
```

```python
y_test = predf1
y_pred = fat_test
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
x_test = pred2f1
print('MAE:', metrics.mean_absolute_error(x_test, y_pred))  
print('MSE:', metrics.mean_squared_error(x_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(x_test, y_pred)))
```

Para a primeira mostra de dados usar o modelo de regresssão linear, ja que o mesmo apresenta erro menor

```python
fat_test2 = df_file2['valor'].values[-1]
df_mfile2 = df_file2.drop(df_file2.index[-1])
fat_test2 = [[fat_test2]]
x2 = pd.DataFrame(df_file2['index'])
y2 = pd.DataFrame(df_file2['valor'])
model1.fit(x2,y2)
model2.fit(x2,y2)
predf2 = (model1.predict([[1541]]))
pred2f2 = (model2.predict([[1541]]))
print('MAE1:', metrics.mean_absolute_error(predf2, fat_test2))  
print('MSE1:', metrics.mean_squared_error(predf2, fat_test2))  
print('RMSE1:', np.sqrt(metrics.mean_squared_error(predf2, fat_test2)))
print('MAE2:', metrics.mean_absolute_error(pred2f2, fat_test2))  
print('MSE2:', metrics.mean_squared_error(pred2f2, fat_test2))  
print('RMSE2:', np.sqrt(metrics.mean_squared_error(pred2f2, fat_test2)))
```
